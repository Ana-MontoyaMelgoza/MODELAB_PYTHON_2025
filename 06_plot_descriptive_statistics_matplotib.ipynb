## Plot histogram of DIC and TA

## Gr√°ficas separadas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_csv(file):
    return pd.read_csv(file)
```

```python
path = "./BD/bd_exercises_curse.csv"

bd_data = read_csv(path)
```

```python
bd_data["dic_micromol_kg"].plot(kind="hist", bins=30, alpha=0.6, color="pink", edgecolor="purple")
plt.title("Histograma de DIC (¬µmol/kg)")
plt.xlabel("DIC (¬µmol/kg)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Histograma de TA
bd_data["ta_micromol_kg"].plot(kind="hist", bins=30, alpha=0.7, color="green", edgecolor="blue")
plt.title("Histograma de TA (¬µmol/kg)")
plt.xlabel("TA (¬µmol/kg)")
plt.ylabel("Frecuencia")
plt.show()
```

## Gr√°ficas Unidas

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# DIC
bd_data["dic_micromol_kg"].plot(kind="hist", bins=30, alpha=0.7, color="pink", edgecolor="purple", ax=axes[0])
axes[0].set_title("Histograma de DIC (¬µmol/kg)")
axes[0].set_xlabel("DIC (¬µmol/kg)")
axes[0].set_ylabel("Frecuencia")

# TA
bd_data["ta_micromol_kg"].plot(kind="hist", bins=30, alpha=0.7, color="green", edgecolor="blue", ax=axes[1])
axes[1].set_title("Histograma de TA (¬µmol/kg)")
axes[1].set_xlabel("TA (¬µmol/kg)")
axes[1].set_ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```
